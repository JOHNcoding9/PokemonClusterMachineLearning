```python
pokemon_names = dataframe["Name"]
pokemon_legendary = dataframe['Legendary']

# Removendo colunas desnecessárias
novo_dataframe = dataframe.drop(columns=['#','Total'], inplace=False)
novo_dataframe.columns
```

```python
# tratamento de nullos
novo_dataframe['Type 2'] = novo_dataframe['Type 2'].fillna('NoType')
novo_dataframe.info()
```

```python
# Contagem de pokemons por tipo 1
counts = novo_dataframe["Type 1"].value_counts()
print(counts)
```

```python
# contagem de pokemons por tipo 2
counts = novo_dataframe["Type 2"].value_counts()
print(counts)
```

```python
# correção de typos (Escrita errada de registros)

# Correção específica para Type 1
novo_dataframe["Type 1"] = (
    novo_dataframe["Type 1"]
        .replace({"ice": "Ice", "bug": "Bug"})
)

# Correção específica para Type 2
novo_dataframe["Type 2"] = (
    novo_dataframe["Type 2"]
        .replace({"ice": "Ice", "bug": "Bug"})
)

counts = novo_dataframe["Type 1"].value_counts()
print(counts)
```

```python
counts = novo_dataframe["Type 2"].value_counts()
print(counts)
```

```python
# Criar um Ordinal encoding que perceba e atribua o mesmo valor para categorias iguais, mas em colunas diferentes. (Fire na coluna Tipo 1 seria a mesam coisa que Fire na coluna Tipo 2.)

# Extrai categorias únicas das duas colunas
todas_categorias = sorted(
    set(novo_dataframe["Type 1"]).union(
        set(novo_dataframe["Type 2"])
    )
)

encoder = OrdinalEncoder(categories=[todas_categorias, todas_categorias])

novo_dataframe[["Type 1", "Type 2"]] = encoder.fit_transform(
    novo_dataframe[["Type 1", "Type 2"]]
)

#encoder.categories_



print(novo_dataframe.head(50))
```

```python
# Escalonamento de dados Z-Socre  (Dados que já passaram pelo Encoder NÃO DEVEM SER ESCALONADOS!)

scaler = StandardScaler()

novo_colunas = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']

novo_dataframe[novo_colunas] = scaler.fit_transform(
    novo_dataframe[novo_colunas]
)

print(novo_dataframe.head(50))
```

```python
# Removendo colunas de tipos para testar modelo de cluster sem dados categóricos
novo_dataframe = novo_dataframe.drop(columns=["Type 1","Type 2"], inplace=False)


print(novo_dataframe.head(50))


novo_dataframe.to_csv("pokemon_tratado.csv", index=False)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report

dataframe = pd.read_csv('Pokemon.csv', index_col=0)

# Replaces typos in rows
dataframe['Type 1'].replace('ice','Ice', inplace=True)
dataframe['Type 1'].replace('bug','Bug', inplace=True)
dataframe['Type 2'].replace('ice','Ice', inplace=True)
dataframe['Type 2'].fillna('NoType', inplace=True) # Replaces NaN with a 'NoType' 
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoded_dataframe = dataframe.copy()
encoder = OrdinalEncoder()

encoder = OrdinalEncoder(categories=[encoded_dataframe['Type 1'].unique()])
encoded_dataframe['Type_1_encoded'] = encoder.fit_transform(encoded_dataframe[['Type 1']])
#encoded_dataframe['Type_1_encoded'].nunique() # Amount of unique values

encoder = OrdinalEncoder(categories=[encoded_dataframe['Type 2'].unique()])
encoded_dataframe['Type_2_encoded'] = encoder.fit_transform(encoded_dataframe[['Type 2']])

encoded_dataframe.drop(['Type 1','Type 2'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = encoded_dataframe.drop(['Name','Generation','Legendary'], axis=1)
y = encoded_dataframe[['Generation']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
#print("\nClassification Report:\n", classification_rep)
```
