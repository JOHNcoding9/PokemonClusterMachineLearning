```python
pokemon_names = dataframe["Name"]
pokemon_legendary = dataframe['Legendary']

# Removendo colunas desnecessárias
novo_dataframe = dataframe.drop(columns=['#','Total'], inplace=False)
novo_dataframe.columns
```

```python
novo_dataframe.isnull().sum()
```

```python
# tratamento de nullos
novo_dataframe['Type 2'] = novo_dataframe['Type 2'].fillna('NoType')
novo_dataframe.info()
```

```python
# Contagem de pokemons por tipo 1
counts = novo_dataframe["Type 1"].value_counts()
print(counts)
```

```python
# contagem de pokemons por tipo 2
counts = novo_dataframe["Type 2"].value_counts()
print(counts)
```

```python
# correção de typos (Escrita errada de registros)

# Correção específica para Type 1
novo_dataframe["Type 1"] = (
    novo_dataframe["Type 1"]
        .replace({"ice": "Ice", "bug": "Bug"})
)

# Correção específica para Type 2
novo_dataframe["Type 2"] = (
    novo_dataframe["Type 2"]
        .replace({"ice": "Ice", "bug": "Bug"})
)

counts = novo_dataframe["Type 1"].value_counts()
print(counts)
```

```python
counts = novo_dataframe["Type 2"].value_counts()
print(counts)
```

```python
# Criar um Ordinal encoding que perceba e atribua o mesmo valor para categorias iguais, mas em colunas diferentes. (Fire na coluna Tipo 1 seria a mesam coisa que Fire na coluna Tipo 2.)

# Extrai categorias únicas das duas colunas
todas_categorias = sorted(
    set(novo_dataframe["Type 1"]).union(
        set(novo_dataframe["Type 2"])
    )
)

encoder = OrdinalEncoder(categories=[todas_categorias, todas_categorias])

novo_dataframe[["Type 1", "Type 2"]] = encoder.fit_transform(
    novo_dataframe[["Type 1", "Type 2"]]
)

#encoder.categories_



print(novo_dataframe.head(50))
```

```python
# Escalonamento de dados Z-Socre  (Dados que já passaram pelo Encoder NÃO DEVEM SER ESCALONADOS!)

scaler = StandardScaler()

novo_colunas = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']

novo_dataframe[novo_colunas] = scaler.fit_transform(
    novo_dataframe[novo_colunas]
)

print(novo_dataframe.head(50))
```

```python
# Removendo colunas de tipos para testar modelo de cluster sem dados categóricos
novo_dataframe = novo_dataframe.drop(columns=["Type 1","Type 2"], inplace=False)


print(novo_dataframe.head(50))


novo_dataframe.to_csv("pokemon_tratado.csv", index=False)
```
